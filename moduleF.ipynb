# 20.1 Problem Statement
Although we can solve derivatives analytically to get an exact solution, sometimes it can be nice to solve it numerically/discretely.
Numerical differentiation will approximate the derivative of a function over a discrete/numerical grid.
Different differentiation schemes will also have different accuracies, which can be compared by looking at the error.

# 20.2 - Finite Difference Approximating Derivatives
In finite deifference approximations of the slope at $x$, we can use different values of the function in the neighborhood of $x$.

## Forward difference
$\frac{f(x_{j+1})-f(x_j)}{x_{j+1}-x_j}$

## Backward difference
$\frac{f(x_{j})-f(x_{j-1})}{x_{j}-x_{j-1}}$
## Central difference
$\frac{f(x_{j+1})-f(x_{j-1})}{x_{j+1}-x_{j-1}}$

## Taylor Series
We can also use Taylor Series to derive an approximation for the derivative of $f$.
## Accuracy
We use Big-O notation to describe the accuracy of the differentiation schemes.
For example the backward difference scheme has an accuracy of $O(h)$, where $h=x_j-x_{j-1}$.

# 20.3 - Approximating of Higher Order Derivatives
We can also use Taylor series to approximate higher order derivatives (e.g., $f''$, $f'''$, etc.).

# 20.4 - Numerical Differentiation with Noise
Sometimes, $f$ might be data points for some values of $x$ as opposed to a proper function.
Thus, there might be noise present in $f$, especially when collecting real world data.

The example below shows an example of the effect of noise in a dataset on its approximated derivative.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

omega = 100
epsilon = 0.01

x = np.arange(0, 2*np.pi, 0.01)
# compute function
y = -2 * np.sin(x)
y_noise = y + epsilon*omega*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
