# Interpolation Problem
To interpolate a dataset means to find an estimation function $\hat{y}(x)$ such that $\hat{y}(x_i)=y_i$ for every point in the data set.
Given a new independent data value $x$ we can also use the function to interpolate its value; in this context $\hat{y}(x)$ is caled an interpolation function.

Unlike regression, we do not need an underlying model for the data.
However, knowing where the data comes from and what it represents might help one select a better interpolation function/algorithm.

```python
import matplotlib.pyplot as plt
# A simple example of a dataset we might want to interpolate for x = 4
x = [1,2,3,5]
y = [2,3,6,10]

plt.plot(x,y,'bo')
plt.show()
```

# Linear Interpolation
Linear interpolation assumes that the estimated point is on the line joining the nearest points to the left and right.

```python
from scipy.interpolate import interp1d

f = interp1d(x,y)
y_hat = f(4)
print(y_hat)
plt.plot(x,y)
plt.plot(x,y,'bo',4,y_hat,'ro') # the estimated value in red
plt.show()
```

# Cubic Spline Interpolation
Unlike linear interpolation, in cubic spline interpolation the interpolating function is a set of piecewise cubic functions.
Specifically, we assume that the points $(x_i,y_i)$ and the subsequent point are joined by a cubic polynomial.
It is important to note that for $n$ points, there are $n-1$ cubic functions.

In Python, we can use scipy's `CubicSpline` to perform cubic spline interpolation.
The `bc_type` argument can be used to set constraints in the spline algorithm.

```python
from scipy.interpolate import CubicSpline
import numpy as np

# use a more interesting dataset
x = [1,2,3,5]
y = [1,4,2,7]
f = CubicSpline(x, y, bc_type="natural")
y_hat_cubic = f(4)
print(y_hat)
plt.plot(np.linspace(1,5,100), f(np.linspace(1,5,100))) # plot the cubic spline

plt.plot(x,y,'bo',4,y_hat_cubic,'ro') # the estimated value in red
plt.show()
```
